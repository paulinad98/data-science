```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```
df_tytanic = pd.read_csv('4. dane_tytanik (train).csv')
df_tytanic
```

```
df_tytanic.info()
```

```
df_tytanic.describe()
```

```
df_tytanic = df_tytanic[['PassengerId',	'Survived',	'Pclass',	'Name',	'Sex', 	'Age',	'SibSp',	'Parch',	'Fare',	'Embarked']]
```

```
df_tytanic = df_tytanic.dropna()
```

```
df_tytanic.describe(include=['object'])
```

```
df_tytanic.describe().T
```

```
df_tytanic.info()
```

```
df_tytanic.columns = []
```

```
sex = pd.DataFrame(columns = ['Male', 'Female'])
sex
```

```
df_tytanic['Male'] = sex.Male
df_tytanic['Female'] = sex.Female
```

```
df_tytanic
```

```
def czyMezczyzna (men):
  if men == 'male':
    return 1
  else:
    return 0

def czyKobieta (women):
  if women == 'female':
    return 1
  else:
    return 0
```

```
df_tytanic['Female'] = df_tytanic.Sex.apply(czyKobieta)
df_tytanic['Male'] = df_tytanic.Sex.apply(czyMezczyzna)
df_tytanic
```

```
df_tytanic ['Embarked_C'] = np.where(df_tytanic.Embarked == 'C', 1, 0)
```

```
df_tytanic
```

```
df_tytanic ['Embarked_S'] = np.where(df_tytanic.Embarked == 'S', 1, 0)
df_tytanic ['Embarked_Q'] = np.where(df_tytanic.Embarked == 'Q', 1, 0)
```

```
df_tytanic
```

```
df_tytanic ['Pclass_1'] = np.where(df_tytanic.Pclass == 1 , 1, 0)
df_tytanic ['Pclass_2'] = np.where(df_tytanic.Pclass == 2 , 1, 0)
df_tytanic ['Pclass_3'] = np.where(df_tytanic.Pclass == 3 , 1, 0)
```

```
df_tytanic
```

```
def tytul (imie):
  return imie.split(', ')[1].split(' ')[0]
```

```
df_tytanic['Title'] = df_tytanic.Name.apply(tytul)
df_tytanic
```

```
df_tytanic = df_tytanic[['PassengerId',	'Survived',	'Pclass_1',	'Pclass_2',	'Pclass_3', 'Age',	'Fare',	'Male',	'Female',	'Embarked_C',	'Embarked_S',	'Embarked_Q']]
df_tytanic
```

```
df_tytanic.describe(include=['object']).T
```

```
df_tytanic.Age = (df_tytanic.Age - np.mean(df_tytanic.Age))/ np.std(df_tytanic.Age)
df_tytanic
```

```
df_tytanic.Fare = (df_tytanic.Fare - np.mean(df_tytanic.Fare))/ np.std(df_tytanic.Fare)
```

```
df_tytanic
```

```
df_tytanic = df_tytanic.set_index(['PassengerId'])
```

```
X = df_tytanic[['Pclass_1',	'Pclass_2',	'Pclass_3', 'Age',	'Fare',	'Male',	'Female',	'Embarked_C',	'Embarked_S',	'Embarked_Q']]
y = df_tytanic[['Survived']]
```

```
X
```

```
y
```

```
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```
X_train.shape
X_test.shape
y_train.shape
y_test.shape
```

```
classifier_rbf=svm.SVC().fit(X_train,y_train)
```

```
classifier_rbf
```

```
classifier_rbf.score(X_train,y_train)
```

```
accuracy_score(y_test,classifier_rbf.predict(X_test))
```
